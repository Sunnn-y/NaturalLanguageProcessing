# BOW 직접 구현

## 실습 2(단어의 순서를 고려하지 않은 BOW)

```python
docs = ['나는 양념 치킨을 좋아해 하지만 후라이드 치킨을 싫어해',
        '나는 후라이드 치킨을 좋아해 하지만 양념 치킨을 싫어해']
```

```python
docs_ls = []

for doc in docs:
  docs_ls.append(doc.split(' '))

docs_ls
```

```python
from collections import defaultdict

word2id = defaultdict(lambda : len(word2id))

for doc in docs_ls:
  for token in doc:
    word2id[token]

word2id
```

```python
import numpy as np
BoW_ls = []
for i, doc in enumerate(docs_ls):
  bow = np.zeros(len(word2id), dtype = int)
  for token in doc:
    bow[word2id[token]] += 1
  BoW_ls.append(bow.tolist())
```

```python
BoW_ls
```

```python
from IPython.core import display as ICD
import pandas as pd
sorted_vocab = sorted((value, key) for key, value in word2id.items())
print('sorted_vocab', sorted_vocab)

vocab = []
for v in sorted_vocab:
  vocab.append(v[1])
print('vocab', vocab)

for i in range(len(docs)):
  print("문서{} : {}".format(i,docs[i]))
  ICD.display(pd.DataFrame([BoW_ls[i]], columns=vocab))
  print("\n\n")
```

# TF-IDF 직접 구현

## 실습

```python
docs = ['오늘 동물원에서 원숭이와 코끼리를 봤어',
        '동물원에서 원숭이에게 바나나를 줬어 바나나를']
```

1. 토큰화 : 공백으로 토큰화 진행

```python
docs_ls = []

for doc in docs:
  docs_ls.append(doc.split(' '))

docs_ls
```

2. 유니크한 토큰 사전 구하기

```python
from collections import defaultdict

word2id = defaultdict(lambda : len(word2id))

for doc in docs_ls:
  for token in doc:
    word2id[token]

dict1 = []
for idx, val in enumerate(word2id):
  dict1.append(val)

word2id
```

3. TF 계산하기

```python
import numpy as np

TF_ls1 = [] # 문서 내 토큰 빈도
TF_ls2 = [] # 문서 내 전체 토큰 빈도
for i, doc in enumerate(docs_ls):
  TF1 = np.zeros(len(word2id), dtype = int)
  TF2 = []
  for token in doc:
    TF1[word2id[token]] += 1
  for freq in range(len(word2id)):
    TF2.append(len(docs_ls[i]))
  TF_ls1.append(TF1.tolist())
  TF_ls2.append(TF2)
```

```python
TF_ls1
```

```python
TF_ls2
```

```python
# 문서 1의 TF
TF1 = []
for m in range(len(TF_ls1[0])):
  TF1.append(TF_ls1[0][m] / TF_ls2[0][m])

import pandas as pd
docu1 = pd.DataFrame({'문서내 토큰 빈도' : TF_ls1[0],
                      '문서내 전체 토큰 빈도' : TF_ls2[0],
                      'TF' : TF1},
                      index = dict1
                    )
docu1
```

```python
# 문서 2의 TF
TF2 = []
for m in range(len(TF_ls1[1])):
  TF2.append(TF_ls1[1][m] / TF_ls2[1][m])

docu2 = pd.DataFrame({'문서내 토큰 빈도' : TF_ls1[1],
                      '문서내 전체 토큰 빈도' : TF_ls2[1],
                      'TF' : TF2},
                       index = dict1
                    )
docu2
```

4. IDF 계산하기

```python
IDF_ls1 = [] # 문서 수
IDF_ls2 = np.zeros(len(word2id), dtype = int) # 문서가 등장한 토큰 수
for doc in word2id:
  IDF_ls1.append(len(docs))

for idx, val in enumerate(word2id):
  for t in dict1:
    if val in docs_ls[0]:
      IDF_ls2[idx] += 1
      break

for idx, val in enumerate(word2id):
  for t in dict1:
    if val in docs_ls[1]:
      IDF_ls2[idx] += 1
      break
```

```python
IDF_ls1
```

```python
IDF_ls2
```

```python
import math
IDF = []
for m in range(len(IDF_ls1)):
  IDF.append(math.log10(IDF_ls1[m]/IDF_ls2[m]))

docu3 = pd.DataFrame({'문서수' : IDF_ls1,
                      '문서가 등장한 토큰 수' : IDF_ls2,
                      'IDF' : IDF},
                     index = dict1
                 )
docu3
```

5. TF-IDF 계산하기

```python
# 문서 1의 TF-IDF
TFIDF1 = []
for m in range(len(TF1)):
  TFIDF1.append(TF1[m] * IDF[m])

TFIDFdocu1 = pd.DataFrame({'TF' : TF1,
                      'IDF' : IDF,
                       'TF-IDF' : TFIDF1},
                     index = dict1
                 )
TFIDFdocu1
```

```python
# 문서 2의 TF-IDF
TFIDF2 = []
for m in range(len(TF2)):
  TFIDF2.append(TF2[m] * IDF[m])

TFIDFdocu2 = pd.DataFrame({'TF' : TF2,
                          'IDF' : IDF,
                          'TF-IDF' : TFIDF2},
                     index = dict1
                 )
TFIDFdocu2
```

## 다른 방식으로 풀이

```python
docs = ['오늘 동물원에서 원숭이와 코끼리를 봤어',
        '동물원에서 원숭이에게 바나나를 줬어 바나나를']
```

1. 토큰화 : 공백으로 토큰화 진행

```python
doc_ls = []

for doc in docs:
  doc_ls.append(doc.split(' '))

doc_ls
```

2. 유니크한 토큰 사전 구하기<br/>
  2-1. 빈 딕셔너리 생성

```python
from collections import defaultdict

word2id = defaultdict(lambda : len(word2id))
word2id
```

2-2. 위에서 생성한 빈 딕셔너리에 유니크한 토큰 넣기

```python
for doc in doc_ls:
  for token in doc:
    word2id[token]
    print(token)
    print('\t',word2id)
word2id
```

3. TF 계산하기<br/>
TF = (특정 단어 등장빈도) / (문서내 전체 등장 단어빈도)

```python
import numpy as np

TDM = np.zeros((len(word2id), len(doc_ls)), dtype=int)

for i, doc in enumerate(doc_ls):
  print(doc)
  for token in doc:
    TDM[word2id[token], i] += 1 # 해당 토큰의 위치(column)
    print(token)
    print(TDM)
    print('\t')
```

```python
# TF = (특정 단어 등장빈도) / (문서내 전체 등장 단어빈도)
def computeTF(TDM):
  doc_len = len(docs) # 문서 개수 2개
  word_len = len(word2id)
  print('문서수 : ',doc_len)
  print('전체 단어수 : ', word_len)

  tf = np.zeros((word_len,doc_len))
  print(tf)

  for doc_i in range(len(doc_ls)):
    print(docs[doc_i])
    for word_i in range(len(word2id)):
      tf[word_i,doc_i] = TDM[word_i,doc_i]/TDM[:,doc_i].sum()
    print(tf)
  return tf
```

```python
tf = computeTF(TDM)
```

```python
tf
```

4. IDF 계산하기<br/>
IDF = -log(단어가 등장한 문서 수 / 총 문서수)

```python
import math
# IDF = -log(단어가 등장한 문서 수 / 총 문서수)
def computeIDF(TDM):
  doc_len = len(docs) # 문서개수 2개
  word_len = len(word2id)
  print('문서수 : ',doc_len)
  print('전체 단어수 : ', word_len)

  idf = np.zeros(word_len)
  print(idf)

  for doc_i in range(len(doc_ls)):
    for word_i in range(len(word2id)):
      idf[word_i] = -math.log10(np.count_nonzero(TDM[word_i,:]) / doc_len)
  print(idf)
  return idf
```

```python
idf = computeIDF(TDM)
idf
```

5. TF-IDF 계산하기

```python
# TF-IDF 곱
def computeTFIDF(TDM):
  tf = computeTF(TDM)
  idf = computeIDF(TDM)
  tfidf = np.zeros(tf.shape)

  for doc_i in range(len(doc_ls)):
    print(docs[doc_i])
    for word_i in range(len(word2id)):
      tfidf[word_i, doc_i] = tf[word_i, doc_i] * idf[word_i]

  print('TF-IDF 최종')
  print(tfidf)
  return tfidf
```

```python
tfidf = computeTFIDF(TDM)
```

```python
tfidf
```

6. 시각화로 확인

```python
import pandas as pd

sorted_vocab = sorted((value, key) for key, value in word2id.items())
vocab = [v[1] for v in sorted_vocab]
print('vocab : ', vocab)
tfidf = computeTFIDF(TDM)
pd.DataFrame(tfidf.T, columns = vocab)
```

## sklearn 으로 풀이
